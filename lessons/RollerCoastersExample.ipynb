# Exploratory Data Analysis - Roller Coasters

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_url = 'https://raw.githubusercontent.com/aoguedao/NISS-workshop/main/data/RollerCoasters.csv'
coasters_df = pd.read_csv(data_url)
coasters_df = coasters_df.dropna()
print(coasters_df.head())
```

```python
#pulls off the drop column of the data frame
all_speeds = coasters_df['Top_Speed']
print(all_speeds)
all_heights = coasters_df['Max_Height']
print(all_heights)
```

```python
plt.hist(all_speeds)
plt.title("Top Speeds")
plt.xlabel ('x')
plt.ylabel ('y')
plt.show()
```

```python
plt.scatter(all_speeds, all_heights)
plt.title("Speed vs Heights Scatter")
plt.xlabel ('Speeds')
plt.ylabel ('Heights')
plt.show()
```

```python
all_coasters = coasters_df['Type']
all_coasters.value_counts().plot(kind = 'pie')
plt.title("Coaster Type")
plt.show()
```
