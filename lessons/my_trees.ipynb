```python
import pandas as pd
```

```python
data_url = 'https://raw.githubusercontent.com/aoguedao/NISS-workshop/main/data/trees.csv'
x = pd.read_csv(data_url)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(x['Girth'])
plt.xlabel("Girth")
```

```python
plt.plot(x['Girth'], x['Volume'],'o', color = 'green')
plt.xlabel("Girth")
plt.ylabel("Volume")
```

```python
plt.plot(x['Height'], x['Volume'],'o', color = 'blue')
plt.xlabel("Height")
plt.ylabel("Volume")
```

```python
plt.plot(x['Girth'], x['Height'],'o', color = 'red')
plt.xlabel("Girth")
plt.ylabel("Height")
```
